# 합성곱 신경망의 구성 요소


## 합성곱
합성곱(convolution)은 마치 입력 데이터에 마법의 도장을 찍어서 유용한 특성만 드러나게 하는 것으로 비유할 수 있습니다.



여기서 중요한 점은 위 그림에 나타난 w1~w3, b가 두 번쪠 합성곱에도 동일하게 적용된다는 점입니다.

합성곱 신경망(convolution neural network, CNN)에서는 완전 연결 신경망과 달리 뉴런을 필터(filter) 또는 커널(kennel)이라고 부릅니다.

입력이 2차원이면 팔터또한 2차원이어야 합니다. 커널 크기는 하이퍼 파라미터 입니다.



왼쪽 위 모서리에서부터 합성곱을 시작하여 오른쪽으로 이동 후 다시 맨 왼쪽 아래로 이동하면서 합성곱을 시행한다.



이렇게 합성곱 계산을 통해 얻은 출력을 특성 맵(feature map)이라고 부릅니다.

밀집층에서 여러 개의 뉴런을 사용하듯이 합성곱 층에서도 여러 개의 필터를 사용합니다.



## 패딩(padding)
입력 배열의 주위를 가상의 원소로 채우는 것을 패딩(padding)이라고 합니다. 실제 입력값이 아니기 때문에 패딩은 0으로 채웁니다.

따라서 계산에 영향을 미치지는 않습니다.

입력과 특성 맵의 크기를 동일하게 만들기 위해 입력 주위에 0으로 패딩 하는 것을 세임 패딩(same padding)이라고 부릅니다.

패딩 없이 순수한 입력 배열에서만 합성곱을 하여 특성 맵을 만드는 경우를 밸리드 패딩(valid padding)이라고 합니다.

패딩은 모서리에 있는 픽셀과 중앙부에 있는 픽셀이 합성곱에 참여하는 비율을 맞추기 위함입니다.



## 스트라이드(strdie)
이동의 크기를 스트라이드(stride)라고 합니다. 기본으로 스트라이드는 1입니다.

## 풀링(pooling)
풀링은 합성곱 층에서 만든 특성 맵의 가로세로 크기를 줄이는 역할을 수행합니다. 하지만 특성 맵의 개수는 줄이지 않습니다.

풀링은 가중치가 없습니다. 영역에서 가장 큰 값을 고르거나 평균값을 계산합니다. 이를 각각 최대 풀링(max pooling)과 평균 풀링(average pooling)이라고 합니다.

풀링에서는 스트라이드와 다르게 겹치지 않고 이동합니다.
