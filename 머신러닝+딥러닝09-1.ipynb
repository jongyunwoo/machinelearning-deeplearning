# 순차 데이터와 순환 신경망


## 순차 데이터
순차 데이터를 다룰 때는 이전에 입력한 데이터를 기억하는 기능이 필요합니다.

다음 샘플을 위해서 이전 데이터가 신경망 층에 순환될 필요가 있습니디.

예를 들어 "I am a boy"는 쉽게 이해할 수 있지만 "boy am a I"는 이해할 수 없습니다.

따라서 데이터를 뒤죽박죽 섞지 않고 순차적으로 사용해야 합니다.

## 순환 신경망
순환 신경망은 완전 연결 신경망에 이전 데이터의 처리 흐름을 순환하는 고리를 추가하면 됩니다.



즉 어떤 샘플을 처리 할 때 바로 이전에 사용했던 데이터를 재사용하는 것입니다.



위 사진처럼 데이터를 처리하는 단계를 타임스텝이라고 합니다.

순환 신경망은 이전 타임스텝의 샘플을 기억하지만 타임스텝이 오래될 수록 순환되는 정보는 희미해 집니다.

순환 신경망에서는 층을 셀이라고 부릅니다. 셀의 출력을 은닉 상태(hidden state)라고 부릅니다.



일반적으로 은닉층의 활성화 함수로는 하이퍼볼릭 탄젠트를 사용합니다.





모든 타임스텝에서 사용되는 가중치는 모두 Wh와 Wx로 동일합니다.

## RNN의 기본구조


합성곱 신경망과 다르게 순환 신경망의 마지막 셀의 출력이 1차원이기 때문에 Flatten 클래스로 펼칠 필요가 없습니다.
